# Data Information

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# load data
data = pd.read_csv('full_data.csv')
```

```python
# show shape of unprocessed data
print('There are', data.shape[0], 'rows')
print('There are', data.shape[1], 'attributes')
```

```python
# get playes by year
seasons = data['Season'].unique()  # Get unique seasons
abs = [data['Season'].value_counts().get(season, 0) for season in seasons]

print(list(seasons))
print(abs)
```

```python
plt.figure(figsize=(10, 6))
plt.bar(seasons, abs, color='skyblue', edgecolor='black')
plt.xlabel('Seasons', fontsize=14)
plt.ylabel('Absolute Counts', fontsize=14)
plt.title('Absolute Counts by Season', fontsize=16)
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('season_abs.png')
plt.show()
```
